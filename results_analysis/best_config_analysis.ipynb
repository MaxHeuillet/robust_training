```python
import pandas as pd

# List of known architecture types
arch_types = [
    'convnext_base',
    'convnext_tiny',
    'deit_small',
    'resnet50',
    'vit_base',
    'vit_small'
]

# Function to extract arch_type from backbone string
def extract_arch_type(backbone):
    for arch in arch_types:
        if arch in backbone:
            return arch
    return 'other'

# Apply to the DataFrame
df_clean['arch_type'] = df_clean['backbone'].apply(extract_arch_type)

# Optional: filter to keep only known architectures
df_clean = df_clean[df_clean['arch_type'] != 'other']
df_clean.to_csv("../figures/hpo_dataset.csv", index=False)
# Compute grouped stats including count
stats = df_clean.groupby(['arch_type', 'loss_function'])[
    ['scheduler', 'lr1', 'lr2', 'weight_decay1', 'weight_decay2',]
].agg(['mean', 'median', 'std', 'count'])

stats.to_csv("../figures/hpo_dataset_analysis.csv", index=True)
```
