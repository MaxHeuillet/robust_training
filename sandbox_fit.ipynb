```python
import numpy as np
from scipy import stats

# Example data (x values and y values)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([100, 80, 65, 50, 35, 25])

# Take the natural logarithm of y
y_log = np.log(y)

# Perform linear regression on the transformed data
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y_log)

# Extract A and B from the linear fit
B = -slope  # Since slope = -B
A = np.exp(intercept)  # Since intercept = ln(A)

print(f"Estimated A: {A}")
print(f"Estimated B: {B}")


# Plotting the original data points
plt.scatter(x, y, label="Data", color="blue")

# Plotting the fitted exponential decay curve
x_fit = np.linspace(min(x), max(x), 100)
y_fit = A * np.exp(-B * x_fit)
plt.plot(x_fit, y_fit, label=f"Fit: y = {A:.2f} * exp({B:.2f} * x)", color="red")

# Adding labels and legend
plt.xlabel("x")
plt.ylabel("y")
plt.title("Exponential Decay Fit")
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data for 10 independent point clouds (x values and y values for each)
np.random.seed(42)  # For reproducibility
n_clouds = 10
x_values = np.linspace(0, 5, 6)  # x values are shared across point clouds

# Generating synthetic decay data with some noise for 10 point clouds
A_true = np.random.uniform(80, 120, n_clouds)  # True A values for each cloud
B_true = np.random.uniform(0.1, 0.5, n_clouds)  # True B values for each cloud
y_values = np.array([A_true[i] * np.exp(-B_true[i] * x_values) + np.random.normal(0, 2, len(x_values)) for i in range(n_clouds)])

# Step 1: Prepare the matrices for linear regression in closed form
# Matrix of x values with a column of ones (for the intercept term)
X = np.vstack([np.ones_like(x_values), -x_values]).T  # X is (6 x 2)

# Prepare the Z matrix (log of y values for each point cloud)
Z = np.log(y_values).T  # Z is (6 x 10), each column is the log of y for a point cloud

# Step 2: Solve for beta using the least squares formula: beta = (X^T X)^(-1) X^T Z
XtX_inv = np.linalg.inv(X.T @ X)  # Compute (X^T X)^(-1)
beta = XtX_inv @ X.T @ Z  # beta is (2 x 10), where first row is C (log(A)) and second row is B

# Extract A and B from beta
C = beta[0, :]  # First row of beta corresponds to C = log(A)
B_estimated = beta[1, :]  # Second row of beta corresponds to -B
A_estimated = np.exp(C)  # Convert log(A) back to A

# Step 3: Plot the results
for i in range(n_clouds):
    x = x_values
    y = y_values[i]
    
    # Plotting the original data points
    plt.scatter(x, y, label=f"Data cloud {i+1}", alpha=0.6)

    # Plotting the fitted exponential decay curve
    x_fit = np.linspace(min(x), max(x), 100)
    y_fit = A_estimated[i] * np.exp(-B_estimated[i] * x_fit)
    plt.plot(x_fit, y_fit, label=f"Fit cloud {i+1}: A={A_estimated[i]:.2f}, B={-B_estimated[i]:.2f}")

# Adding labels and legend
plt.xlabel("x")
plt.ylabel("y")
# plt.yscale('log')  # Set logarithmic scale on y-axis for better visualization
plt.title("Exponential Decay Fit for 10 Point Clouds using Matrix Computations")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()

# Display the plot
plt.show()

# Display the estimated A and B values for each cloud
for i in range(n_clouds):
    print(f"Point Cloud {i+1}: Estimated A = {A_estimated[i]:.2f}, Estimated B = {-B_estimated[i]:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

# Function to fit one exponential decay curve
import numpy as np

def fit_one_curve(y):
    # Ensure y has no zeros to avoid log(0) which would cause -inf
    # y = np.clip(y, a_min=1e-10, a_max=None)

    # Step 1: Compute z = ln(y)
    z = np.log(y)

    # x is a fixed array based on length of y, no need to recompute its sum multiple times
    n = len(y)
    x = np.arange(n)
    
    # Step 2: Compute the necessary sums for linear regression
    S_x = (n - 1) * n / 2  # sum of 0 to n-1
    S_xx = (n - 1) * n * (2 * (n - 1) + 1) / 6  # sum of squares of 0 to n-1
    S_z = np.sum(z)
    S_xz = np.dot(x, z)  # dot product of x and z is the same as sum(x * z)

    # Step 3: Compute the slope (for -B)
    denominator = n * S_xx - S_x ** 2
    slope = (n * S_xz - S_x * S_z) / denominator
    B = -slope  # Ensuring B is positive for decay

    # Step 4: Compute the intercept (for ln(A))
    intercept = (S_z - slope * S_x) / n

    # Step 5: Compute A from the intercept
    A = np.exp(intercept)  # A = e^C

    return A, B, intercept


# Example data for 60,000 independent point clouds (x values and y values for each)
np.random.seed(42)  # For reproducibility
n_clouds = 60000
x_values = np.linspace(0, 5, 6)  # x values are shared across point clouds

# Generating synthetic decay data with some noise for 60,000 point clouds
A_true = np.random.uniform(80, 120, n_clouds)  # True A values for each cloud
B_true = np.random.uniform(0.1, 0.5, n_clouds)  # True B values for each cloud
values = np.array([A_true[i] * np.exp(-B_true[i] * x_values) + np.random.normal(0, 2, len(x_values)) for i in range(n_clouds)])

# Perform fitting for each point cloud
solutions = [fit_one_curve(value) for value in tqdm(values)]

# Plot a subset of results for visualization
n_plots = 5  # Number of plots to show
for i in range(n_plots):
    x = x_values
    y = values[i]
    A, B, _ = solutions[i]

    # Plot the original data points
    plt.scatter(x, y, label=f"Data cloud {i+1}", alpha=0.6)

    # Plot the fitted exponential decay curve
    x_fit = np.linspace(min(x), max(x), 100)
    y_fit = A * np.exp(-B * x_fit)
    plt.plot(x_fit, y_fit, label=f"Fit cloud {i+1}: A={A:.2f}, B={B:.2f}")

# Adding labels and legend
plt.xlabel("x")
plt.ylabel("y")
# plt.yscale('log')  # Set logarithmic scale on y-axis for better visualization
plt.title("Exponential Decay Fit for a Subset of Point Clouds")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()

# Show the plot
plt.show()
```
